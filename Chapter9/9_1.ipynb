## 순차데이터
스크린샷 2026-04-06 09.05.53.png

순차데이터는 텍스트나 시계열 데이터와 같이 순서에 의미가 있는 데이터를 말한다.

이전 장에서 배웠던 완전 연결 신경망과 합성곱 신경망은 모두 입력 데이터의 흐름이 앞으로만 전달되는 피드포워드 신경망이라고 부른다.

순차데이터처럼 신경망이 이전에 처리했던 샘플을 다음 샘플을 처리하는데 재사용하기 위해서는 데이터 흐름이 피드포워드 신경망처럼 앞으로만 가면 곤란한데 다음 샘플을 위해서 이전 데이터가 신경망 층에 순환도리 필요가 있다.

이런 신경망이 순환 신경망이다.

## 순환 신경망
스크린샷 2026-04-06 09.19.02.png

순환 신경망은 일반적인 완전 연결 신경망에서 그림과 같이 붉은 고리가 생긴 형태다.

뉴런의 출력이 다시 자기 자시능로 전달되는데 바로 이전에 사용했던 데이터를 재사용하는 셈이다.

두번째 그림처럼 결국에 Oc에는 앞선 두 정보가 포함되어 있는데 이렇게 샘플을 처리하는 한 단계를 `타임스텝`이라고 부른다.

순환 신경망에서는 층을 `셀`이라고 부른다. 한 셀에는 여러 개의 뉴런이 있지만 뉴런을 모두 표시하지 않고 하나의 셀로 층을 표현한다. 또한 셀의 출력을 `은닉상태`라고 부른다.

은닉층의 활성화 함수로는 tanh 가 많이 사용된다.

완전 연결되기에 가중치는 12개이다. 순환층에서 타임스텝에 재사용 되는 은닉 상태를 위한 가중치 Wh의 크기는?

r1~3뉴런들은 데이터를 받고 자기 자신만 순환하는게 아니라 나머지 뉴런들에게도 전달된다. 즉 이전 타임스텝의 은닉 상태는 다음 타임스텝의 뉴런에 완전히 연결 된다.

따라서 이 순환층에서 은닉 상태를 위한 가중치는 3*3 = 9개이다.

이런 뉴런들을 모은것들을 그림으로 전부 표현하기 힘드니 하나의 `셀`로 부름
